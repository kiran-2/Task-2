# Objective:-
                                              in this regression task we will predict the percentage of marks that a student is expected to score based upon no of hours they studied.This is  a simple linear regression task as it involves just two variables.

# importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

# Reading Data

```python
score_data=pd.read_csv('http://bit.ly/w-data')
score_data
print('data imported successfully')
score_data.head()
```

# Data Exploration

```python
score_data.shape
```

```python
score_data.describe()
```

```python
x=score_data.iloc[:,:-1].values
y=score_data.iloc[:,1].values
x
y
```

# Train_Test_Split Model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=2)
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
t=reg.score(x_test,y_test)
t*100
```

```python
print("training is completed")
```

```python
print("intercept is:",reg.intercept_)
```

```python
print("coefficient is",reg.coef_)
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
df=pd.DataFrame({'actual':y_test,'predicted':y_pred})
df
```
